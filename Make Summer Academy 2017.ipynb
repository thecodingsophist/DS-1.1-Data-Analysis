## Next, are there any trends in each of the tracks in New York City?

```python
df_NY_Apps = df_NY[(df_NY['Track'] == "Apps")]
```

```python
ax = df_NY_Apps.plot.line(x='Week', y='Schedule Pacing', figsize=(20,10))
```

## It looks like people consistently felt that the class was too fast, with a few people feeling like the class was too slow, until week 4, when class was became too fast for pretty much everyone

```python
df_NY_Apps_Explorer = df_NY[(df_NY['Track'] == "Apps, Explorer")]
```
