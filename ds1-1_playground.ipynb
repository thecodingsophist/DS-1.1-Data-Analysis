## How many Roses and Jacks were on the _Titanic_?

- Jack (Age 20, Sex Male, Class 3) (Survived 0)
- Rose (Age 17, Sex Female, Class 1) (Survived 1)

```python
rose_age_mask = df["Age"] == 17
rose_sex_mask = df["Sex"] == "female"
rose_class_mask = df["Pclass"] == 1
rose_survival_mask = df["Survived"] == 1

jack_age_mask = df["Age"] == 20
jack_sex_mask = df["Sex"] == "male"
jack_class_mask = df["Pclass"] == 3
jack_survival_mask = df["Survived"] == 0
```

```python
rose_df = df[(rose_age_mask) & (rose_sex_mask) & (rose_class_mask) & (rose_survival_mask)]
jack_df = df[(jack_age_mask) & (jack_sex_mask) & (jack_class_mask) & (jack_survival_mask)]
len(jack_df), len(rose_df)
```
